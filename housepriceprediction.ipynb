```python
# Explore values with missing values
missing_values = df[df.isnull().any(axis = 1)]
missing_values
```

```python
df_null = df.isnull().sum()
df_null = df_null[df_null > 0]
df_null

missing_percentage = (df_null / len(df)) * 100
missing_percentage
```

```python
# Summary statistics of categorical columns
categorical_features = df.select_dtypes(include = [object])
categorical_features.describe().T
```

```python
import plotly.graph_objects as go
```

```python
import scipy.stats as stats

# Fit a normal distribution to the SalePrice data
mu, sigma = stats.norm.fit(df['SalePrice'])

# Create a histogram of the SalePrice column
hist_data = go.Histogram(x=df['SalePrice'], nbinsx=50, name="Histogram", opacity=0.75, histnorm='probability density', marker=dict(color='purple'))

# Calculate the normal distribution based on the fitted parameters
x_norm = np.linspace(df['SalePrice'].min(), df['SalePrice'].max(), 100)
y_norm = stats.norm.pdf(x_norm, mu, sigma)

# Create the normal distribution overlay
norm_data = go.Scatter(x=x_norm, y=y_norm, mode="lines", name=f"Normal dist. (μ={mu:.2f}, σ={sigma:.2f})", line=dict(color="green"))

# Combine the histogram and the overlay
fig = go.Figure(data=[hist_data, norm_data])

fig.show()
```

```python
# Distribution of Dwelling types and Their Relationship with Sales Price

dwelling_types = df['BldgType'].value_counts()
dwelling_price = df.groupby('BldgType')['SalePrice'].mean()
formatted_dwelling_prices = ['$' + f"{value:,.2f}" for value in dwelling_price.values]

fig = px.bar(x = dwelling_types.index,y = dwelling_types.values,text = dwelling_types.values)

# Customize the layout
fig.update_traces(textposition='outside')
fig.update_xaxes(title = 'Building Types')
fig.update_yaxes(title = 'Count')
fig.update_layout(title = 'Distribution of Building Types')

fig.show()
```

```python
df_fig = pd.DataFrame()

df_fig ['BldgType'] =''
df_fig ['SalePrice'] =''
df_fig ['FormattedPrice'] =''

dwelling_price = df.groupby('BldgType')['SalePrice'].mean()

# Format and print the values
for bldg_type, price in dwelling_price.items():
    formatted_price = "${:,.2f}".format(price)
    dwelling_price['bldg_type'] = formatted_price
    new_row = {'BldgType':bldg_type,'SalePrice':price,'FormattedPrice':formatted_price}
    #df_fig = df_fig.append(new_row,ignore_index=True)
    df_fig_append = pd.DataFrame([new_row])
    df_fig = pd.concat([df_fig,df_fig_append],ignore_index=True)

fig2 = px.bar(df_fig, x ='BldgType', y = 'SalePrice' , text ='FormattedPrice')
fig2.update_layout(title = 'Sale Prices by Building Type', plot_bgcolor='black',paper_bgcolor='black', font_color = 'white',bargap =0.4)


fig2.show()
```
